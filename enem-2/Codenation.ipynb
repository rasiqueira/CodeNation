```python
test = teste.drop(['NU_INSCRICAO', 'CO_UF_RESIDENCIA', 'SG_UF_RESIDENCIA', 'NU_IDADE',
       'TP_SEXO', 'TP_COR_RACA', 'TP_NACIONALIDADE', 'TP_ST_CONCLUSAO',
       'TP_ANO_CONCLUIU', 'TP_ESCOLA', 'TP_ENSINO', 'IN_TREINEIRO',
       'TP_DEPENDENCIA_ADM_ESC', 'IN_BAIXA_VISAO', 'IN_CEGUEIRA', 'IN_SURDEZ',
       'IN_DISLEXIA', 'IN_DISCALCULIA', 'IN_SABATISTA', 'IN_GESTANTE',
       'IN_IDOSO', 'TP_PRESENCA_CN', 'TP_PRESENCA_CH', 'TP_PRESENCA_LC','CO_PROVA_CN', 'CO_PROVA_CH', 'CO_PROVA_LC'], axis=1)
```

```python
df1 = df[['CO_PROVA_MT', 'NU_NOTA_CN', 'NU_NOTA_CH', 'NU_NOTA_LC', 'TP_LINGUA',
       'TP_STATUS_REDACAO', 'NU_NOTA_COMP1', 'NU_NOTA_COMP2', 'NU_NOTA_COMP3',
       'NU_NOTA_COMP4', 'NU_NOTA_COMP5', 'NU_NOTA_REDACAO', 'Q001', 'Q002',
       'Q006', 'Q024', 'Q025', 'Q026', 'Q027', 'Q047']]
```

```python
df1
```

```python
data = [df1, test]
for dataset in data:
    dataset[['CO_PROVA_MT', 'NU_NOTA_CN', 'NU_NOTA_CH', 'NU_NOTA_LC', 'TP_LINGUA',
       'TP_STATUS_REDACAO', 'NU_NOTA_COMP1', 'NU_NOTA_COMP2', 'NU_NOTA_COMP3',
       'NU_NOTA_COMP4', 'NU_NOTA_COMP5', 'NU_NOTA_REDACAO', 'Q001', 'Q002',
       'Q006', 'Q024', 'Q025', 'Q026', 'Q027', 'Q047']] = dataset[['CO_PROVA_MT', 'NU_NOTA_CN', 'NU_NOTA_CH', 'NU_NOTA_LC', 'TP_LINGUA',
       'TP_STATUS_REDACAO', 'NU_NOTA_COMP1', 'NU_NOTA_COMP2', 'NU_NOTA_COMP3',
       'NU_NOTA_COMP4', 'NU_NOTA_COMP5', 'NU_NOTA_REDACAO', 'Q001', 'Q002',
       'Q006', 'Q024', 'Q025', 'Q026', 'Q027', 'Q047']].fillna(-1)
```

```python
df.head()
```

```python
categoria = ['CO_PROVA_MT', 'Q001', 'Q002',
       'Q006', 'Q024', 'Q025', 'Q026', 'Q027', 'Q047']
data = [df1, test]
for dataset in data:
    for cat in categoria:
        dataset[cat] = dataset[cat].astype('category')        
```

```python
# Get dummies
df1 = pd.get_dummies(df1)
test = pd.get_dummies(test)
```

```python
df1.head()
```

```python
df1.shape
```

```python
# Criando objetos
df['NU_NOTA_MT'] = df['NU_NOTA_MT'].fillna(-1)
X_train = df1.values
Y_train = df['NU_NOTA_MT'].values
X_test = test.values
```

```python
Y_train
```

## Construindo e treinando o modelo

```python
# Utilizando um classificador Naive Bayes
from sklearn.linear_model import LinearRegression
```

```python
# Criando o modelo preditivo
modelo_v1 = LinearRegression()
```

```python
# Treinando o modelo
modelo_v1.fit(X_train, Y_train)
```

## Fazendo Previsões Com o Modelo Treinado

```python
submission =pd.DataFrame()
submission['NU_INSCRICAO'] = teste['NU_INSCRICAO']
submission['NU_NOTA_MT'] = modelo_v1.predict(X_test)
```

```python
submission.to_csv('answer.csv',index = False)
```
